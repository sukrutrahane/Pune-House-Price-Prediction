```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
```

```python
house=pd.read_csv('clean_data.csv')
house
```

```python
y = house['price']
x = house[['size', 'total_sqft', 'bath', 'balcony', 'site_location']]
```

```python
features = ['size', 'total_sqft', 'bath', 'balcony', 'site_location']
target_variable = 'price'
```

```python
column_trans = make_column_transformer((OneHotEncoder(sparse=False), ['site_location']), remainder='passthrough')
```

```python
# Create and train a linear regression model
model = LinearRegression()
model.fit(column_trans.fit_transform(house[features]), house[target_variable])
```

```python
def predict(size, total_sqft, bath, balcony, site_location):
    input_data = pd.DataFrame({'size': [size], 'total_sqft': [total_sqft], 'bath': [bath], 'balcony': [bath], 'site_location': [site_location]})
    input_data_transformed = column_trans.transform(input_data)
    
    predicted_price = model.predict(input_data_transformed)[0]*100000
    return predicted_price
```

```python
predicted_price = predict(size=2, total_sqft=1200, bath=3, balcony=1, site_location='Koregaon Park')
print(f"Predicted Price: Rs. {predicted_price:.2f}")
```
